```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('data/00-puppy.jpg')
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
```

```python
# red channel values 0-255, 0 = no red, pure black = 255 full pure red
plt.imshow(pic_red[:,:,0], cmap='gray')
```

```python
# green channel
plt.imshow(pic_red[:,:,1], cmap='gray')
```

```python
# blue channel
plt.imshow(pic_red[:,:,2], cmap='gray')
```

```python
# green channel remove
pic_red[:,:,1] = 0
# blue channel remove
pic_red[:,:,2] = 0
plt.imshow(pic_red)
```
